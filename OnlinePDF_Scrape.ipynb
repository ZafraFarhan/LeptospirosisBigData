```python
import requests
from bs4 import BeautifulSoup
import os
import pdfplumber
import pandas as pd

# Define the URL of the webpage
url = 'https://www.epid.gov.lk/weekly-epidemiological-report/weekly-epidemiological-report'

# Define the directory to save downloaded PDFs
output_dir = 'pdfs'
os.makedirs(output_dir, exist_ok=True)

def download_pdfs(url, output_dir):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all accordions
    accordions = soup.find_all(class_='accordions')

    for accordion in accordions:
        # Find all content within the accordion
        content = accordion.find(class_='content')
        if content:
            # Find all product links within the content
            products = content.find_all('li', class_='product')
            for product in products:
                a_tag = product.find('a', href=True)
                if a_tag:
                    pdf_url = a_tag['href']
                    pdf_name = os.path.basename(pdf_url)
                    pdf_path = os.path.join(output_dir, pdf_name)

                    # Download the PDF
                    pdf_response = requests.get(pdf_url)
                    with open(pdf_path, 'wb') as file:
                        file.write(pdf_response.content)

                    print(f'Downloaded: {pdf_path}')
    return [os.path.join(output_dir, f) for f in os.listdir(output_dir) if f.endswith('.pdf')]

pdf_paths = download_pdfs(url, output_dir)
```

```python
import pdfplumber
import pytesseract
from PIL import Image
import pandas as pd

def extract_table_from_pdf(pdf_path, table_title):
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Check for table title in the page text
            if table_title in page.extract_text():
                # Extract tables from this page
                tables = page.extract_tables()
                for table in tables:
                    df = pd.DataFrame(table[1:], columns=table[0])
                    return df
    return None

def is_rotated(image_path):
    """Check if the image is rotated using OCR."""
    text = pytesseract.image_to_string(Image.open(image_path))
    # Assuming that if text has more spaces than words, it might be rotated
    words = text.split()
    spaces = text.count(' ')
    return spaces > len(words) * 1.5

def correct_rotation(image_path):
    """Correct the rotation of the image and return the corrected image."""
    image = Image.open(image_path)
    for angle in [0, 90, 180, 270]:
        rotated_image = image.rotate(angle, expand=True)
        if not is_rotated(rotated_image):
            return rotated_image
    return image  # Return original image if no correct rotation is found

def extract_table_from_image(image):
    """Extract table from an image using OCR and return as DataFrame."""
    text = pytesseract.image_to_string(image)
    data = [line.split() for line in text.split('\n') if line.strip()]
    df = pd.DataFrame(data[1:], columns=data[0])
    return df

def extract_table_from_pdf_page(pdf_path, page):
    """Extract table from a specific page of the PDF."""
    # Save the page as an image
    image_path = "temp_page.png"
    page.to_image().save(image_path)

    # Check if the image is rotated and correct it if needed
    if is_rotated(image_path):
        corrected_image = correct_rotation(image_path)
        table_df = extract_table_from_image(corrected_image)
    else:
        table_df = extract_table_from_image(Image.open(image_path))

    return table_df

def extract_table_from_pdf_updated(pdf_path, table_title):
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Check for table title in the page text
            if table_title in page.extract_text():
                table_df = extract_table_from_pdf_page(pdf_path, page)
                if table_df is not None:
                    return table_df
    return None

def concatenate_tables(pdf_paths):
    concatenated_df = pd.DataFrame()
    problematic_pdfs = ['pdfs/vol_43_no_46-english.pdf', 'pdfs/vol_43_no_35-english.pdf']

    for pdf_path in pdf_paths:
        if pdf_path in problematic_pdfs:
            print(f'Skipping problematic PDF: {pdf_path}')
            continue
        table_df = extract_table_from_pdf_updated(pdf_path, table_title)
        if table_df is not None:
            concatenated_df = pd.concat([concatenated_df, table_df], ignore_index=True)

    return concatenated_df

# Example usage:
final_df = concatenate_tables(pdf_paths)
print(final_df)
```

```python
import pdfplumber
import pandas as pd
from PIL import Image
import io

def extract_and_correct_table_from_pdf(pdf_path, table_title):
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Check for table title in the page text
            if table_title in page.extract_text():
                # Get the rotation angle of the page
                rotation = page.rotation

                if rotation != 0:
                    # Rotate the page back to 0 degrees
                    pil_image = page.to_image()
                    pil_image = pil_image.rotate(-rotation, expand=True)
                    image_bytes = pil_image.to_bytes()

                    # Use pdfplumber's image functionality to work with the rotated image
                    page = pdfplumber.open(io.BytesIO(image_bytes)).pages[0]

                # Extract tables from the page
                tables = page.extract_tables()
                for table in tables:
                    # Convert the table to DataFrame
                    df = pd.DataFrame(table[1:], columns=table[0])
                    return df
    return None

# Path to the PDF file
pdf_path = 'pdfs/vol_42_no_20-english.pdf'  # Update this with your actual PDF file name
table_title = 'Selected notifiable diseases reported by Medical Officers of Health'

# Extract the table
df = extract_and_correct_table_from_pdf(pdf_path, table_title)

if df is not None:
    print(df)
else:
    print("Table not found.")
```

```python
import requests
from bs4 import BeautifulSoup
import os
import pdfplumber
import pandas as pd

# Define the URL of the webpage
url = 'https://www.epid.gov.lk/weekly-epidemiological-report/weekly-epidemiological-report'

# Define the directory to save downloaded PDFs
output_dir = 'pdfs'
os.makedirs(output_dir, exist_ok=True)


def extract_table_from_pdf(pdf_path, table_title):
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Check for table title in the page text
            if table_title in page.extract_text():
                # Extract tables from this page
                tables = page.extract_tables()
                for table in tables:
                    # Convert the table to DataFrame
                    df = pd.DataFrame(table[1:], columns=table[0])
                    return df
                # Handle rotated tables
                rotated_page = page.rotate(90)
                if table_title in rotated_page.extract_text():
                    tables = rotated_page.extract_tables()
                    for table in tables:
                        # Convert the table to DataFrame
                        df = pd.DataFrame(table[1:], columns=table[0])
                        return df
    return None

# Path to the PDF file
table_title = 'Selected notifiable diseases reported by Medical Officers of Health'

def concatenate_tables(pdf_paths):
    # Initialize an empty DataFrame to store concatenated tables
    concatenated_df = pd.DataFrame()

    # List of problematic PDFs to ignore
    problematic_pdfs = [
        'pdfs/vol_43_no_46-english.pdf',
        'pdfs/vol_43_no_35-english.pdf'
    ]

    for pdf_path in pdf_paths:
        if pdf_path in problematic_pdfs:
            print(f'Skipping problematic PDF: {pdf_path}')
            continue
        table_df = extract_table_from_pdf(pdf_path, table_title)
        if table_df is not None:
            concatenated_df = pd.concat([concatenated_df, table_df], ignore_index=True)

    return concatenated_df

# Execute the functions
final_df = concatenate_tables(pdf_paths)

# Optionally, save the final DataFrame to a CSV file
final_df.to_csv('combined_tables.csv', index=False)
print('Tables combined and saved to combined_tables.csv')
```
